**FINDING AVERAGE RATING**

```python
s=0
for i in df['Rating']:
  s+=i
s=int(s)
print(s/len(df['Rating']))
```

**APPS WITH RATING OF 5.0**

```python
c=0
for i in df['Rating']:
  if(i==5.0):
    c=c+1
print("There are",c,"apps with rating of 5.0")
```

```python
c=0
for i in df['Rating']:
  if(4<=i<=4.5):
    c=c+1
print("There are",c,"apps with rating between 4.0 and 4.5")
```

**AVERAGE NUMBER OF REVIEWS**

```python
s=0
for i in df['Reviews']:
  s+=int(i);
s=int(s)
print("Average number of reveiws are",int(s/len(df['Reviews'])))
```

**CLEANING DATA**

```python
import numpy as np
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
data.iloc[:,2:3]=imputer.fit_transform(data.iloc[:,2:3])
data=data.dropna()
print(data.isnull().sum())
```

**Q1. How many free apps are there in ART_AND_DESIGN?**

```python
df=data.values
count=0
for i in df:
  if i[1]=='ART_AND_DESIGN' and i[6]=='Free':
    count=count+1
print(count)
```

**Q2  List all the free apps with rating more then 4.5 and category is FAMILY?**

```python

df_pr=data[(data['Type']=='Free') &  (data['Rating']>4.5) & (data['Category']=='FAMILY')]
result=df_pr.values
for i in result:
  print(i[0])
```
